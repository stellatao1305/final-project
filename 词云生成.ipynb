```python
import jieba  #结巴分词
from wordcloud import WordCloud  #词云展示库
from PIL import Image  #图像处理库
import numpy as np  #支持多维数组和矩阵运算
import matplotlib.pyplot as plt 
```

```python
with open('/Users/tsubomikaji/Desktop/弹幕/清理data(2版/clean3.txt','r',encoding="UTF-8") as f:
    file = f.read()
```

```python
data_cut =jieba.cut(file,cut_all = False)
```

```python
stop_words = [] #创建空列表
with open("/Users/tsubomikaji/Downloads/stopwords-master/cn_stopwords.txt", 'r', encoding='utf-8') as f:
    for line in f:
        if len(line)>0:
            stop_words.append(line.strip()) 
data_result = [i for i in data_cut if i not in stop_words] 
print(data_result)
```

```python
text = " ".join(data_result).replace("\n","") #连接成字符串
print(text)
```

```python
wc = WordCloud(
    #设置字体，不指定就会出现乱码,这个字体文件需要下载
    font_path = "/System/Library/Fonts/PingFang.ttc",
    width=1000,
    height=800,
    background_color = "white",
    max_words = 300,
)
```

```python
wc.generate(text)

 # 保存词云图
wc.to_file("/Users/tsubomikaji/Downloads/heitu3.jpg") #保存图片
 
 # 展示
plt.imshow(wc) #对图片进行处理，并显示其格式
plt.axis("off") #关闭坐标轴
plt.show() #将图片显示出来
```
